### Pandas Datatype: Series vs Dataframe

[Pandas 자료](https://choieungi-project.notion.site/pandas-1c3a7a8f2b3a44cf90bb04bf75571230)

**Series**

- 하나의 데이터 타입(`dtype`)을 갖고 label을 갖는 일차원 배열 <br><br>

**Dataframe**

- 여러 데이터 타입을 갖는 이차원 라벨 자료 구조
- 여러 Series로 구성된 것으로도 볼 수 있음
- Dataframe에서 하나의 Column으로 접근할 때 Series로 변환된다








```python
# pd.read_csv를 이용하면 기본적으로 csv를 읽어 Dataframe으로 만들어준다.
data = pd.read_csv('./sample_data/WorldCups.csv')
type(data)
```

```python
# Dataframe에서 직접 값으로 접근하면 Series로 리턴
a = data[['Year', 'Country']]
print(type(a))
a
```

```python
# Dataframe에서 배열 인덱스로 접근하면 Dataframe 타입으로 리턴
a = data[['Year']]
print(type(a))
a
```

### pandas Groupby
[공식문서](https://pandas.pydata.org/docs/user_guide/groupby.html)

Groupby()는 ① 전체 데이터를 그룹별로 분할(split)하고, ② mean(), sum(), count()와 같은 Aggregate function을 사용하여 연산(apply)하고, 연산 결과를  ③ 다시 합치는(combine) 과정을 거치게 됩니다.

```python
df = pd.DataFrame(
    [
        ("bird", "Falconiformes", 389.0, 3),
        ("bird", "Psittaciformes", 24.0, 4),
        ("mammal", "Carnivora", 80.2, 5),
        ("mammal", "Primates", np.nan, 100),
        ("mammal", "Carnivora", 58, 20),
     
    ],
    index=["falcon", "parrot", "lion", "monkey", "leopard"],
    columns=("class", "order", "max_speed", "age"),
)
df
```

```python
grouped = df.groupby(["class"])  #1. Split
ret = grouped.sum() # 2. aggregate function 
ret
```

**concat**

- Dataframe을  합치는 메소드
- `axis` property가 0이면 아래 위로 합치고, 1이면 왼쪽 오른쪽으로 합친다. (default = 0)
    - union, join
- `sort` 인덱스 기준으로 정렬할 것인지에 대한 property (default = False)

[pandas.concat - pandas 1.4.2 documentation](https://pandas.pydata.org/docs/reference/api/pandas.concat.html)

**fillna(value)**
[공식문서](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html)

> Fill NA/NaN values using the specified method.
> 
- dataframe의 Na, NaN값을 value로 채워줌

**sort_values()**

- Series를 값을 기준으로 정렬
- `ascending=False` 속성을 통해 Series를 내림차순으로 정렬 가능

**drop()**


```python
df = pd.DataFrame(
    [
        ("bird", "Falconiformes", 389.0, 3),
        ("bird", "Psittaciformes", 24.0, 4),
        ("mammal", "Carnivora", 80.2, 5),
        ("mammal", "Primates", np.nan, 100),
        ("mammal", "Carnivora", 58, 20),
     
    ],
    index=["falcon", "parrot", "lion", "monkey", "leopard"],
    columns=("class", "order", "max_speed", "age"),
)
df
```

```python
max_speed_df = df.groupby('class')['max_speed'].mean()
age_df = df.groupby('class')['age'].mean()
print(max_speed_df, age_df)
```

```python
ret = pd.concat([max_speed_df, age_df], axis=0) #0은 위 아래로 합친다
ret
```

```python
ret = pd.concat([max_speed_df, age_df], axis=1) #0은 위 아래로 합친다
ret
```

```python
# fillna
print(df)
df = df.fillna(0)
df
```

```python
#sort_values
max_speed_series = df['max_speed']
print(max_speed_series)
#ascending(오름차순)
max_speed_series = max_speed_series.sort_values()
max_speed_series
```

```python
#descending(내림차순)
max_speed_series = max_speed_series.sort_values(ascending=False)
max_speed_series
```

```python
#drop
ret = pd.concat([max_speed_df, age_df], axis=1) #0은 위 아래로 합친다
ret = ret.drop(['age'], axis=1)
ret
```
