### 2-3. Summary Fuctions and Maps

```python
reviews.points.describe()
```

```python
reviews.points.mean()
```

```python
reviews.taster_name.unique()
```

```python
reviews.taster_name.value_counts()
```

```python
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns').head()
```

```python
reviews.head(1)
```

```python
review_points_mean = reviews.points.mean()
reviews.points - review_points_mean
```

### 2-4. Grouping and Sorting

```python
reviews.groupby('points').points.count()
```

```python
reviews.groupby('points').price.min()
```

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
reviews.groupby(['country']).price.agg([len, min, max])
```

```python
reviews.sort_values(by='country', ascending = False).head()
```

```python
reviews.sort_values(by=['country', 'points'], ascending = False).head()
```
